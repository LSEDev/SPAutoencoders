```python
import pandas as pd
import requests
import datetime
from pandas_datareader import data, wb
import os
import sys

from CompanyHolder import CompanyHolder

SYMBOLS_PATH = 'Symbols'
STOCK_DF_PATH = 'stock_df'
api_key = 'eb6e201c3cebcfcb6c7596eaf6bbf2d39d685ae8'
```

```python
# First update the S&P 500 companies
# Note that this can change over time
if not os.path.exists(SYMBOLS_PATH):
    os.makedirs(SYMBOLS_PATH)
    
# CompanyHolder(SYMBOLS_PATH).save_company_names(reload=False)
```

```python
def save_stock_market_data(start):
    '''Save a variety of metrics for stocks on S&P500 into csv files'''
    
    CH = CompanyHolder(SYMBOLS_PATH)
    #set the end time to today
    end = datetime.datetime.now()
    problem_list = [] # shares that the API doesn't return 
    
    if not os.path.exists(STOCK_DF_PATH + '/Individual_Companies'):
        os.makedirs(STOCK_DF_PATH + '/Individual_Companies')
    
    #get all the data and store it in a dictionary
    for ind, comp in enumerate(CH.Symbols):
        if not os.path.exists(STOCK_DF_PATH + '/Individual_Companies/{}_StockData.csv'.format(comp)):
            
            try:
                df_prices = data.DataReader(comp.replace('.','-'), 'tiingo', start, end,
                                    api_key=api_key)



                requestResponse = requests.get("https://api.tiingo.com/tiingo/fundamentals/" + comp.replace('.','-') + "/daily?token=" + api_key, 
                                                                 headers={'Content-Type': 'application/json'}, 
                                                                 params={'startDate': start, 'endDate': end})
                df_fundamentals = pd.DataFrame(requestResponse.json())
                df_fundamentals['date']=pd.to_datetime(df_fundamentals['date'])

                df = df_prices.merge(df_fundamentals, on='date') 

                df.to_csv(STOCK_DF_PATH + '/Individual_Companies/{}_StockData.csv'.format(comp))

                print("Retrieved {}".format(comp))
                
            except:
                problem_list.append(comp.replace('.','-'))
                print('\nCould not get', comp.replace('.','-'), '\n')
            
        else:
            print("Already have {}".format(comp.replace('.','-')))
            
    print("These stocks were not found:", problem_list)
            
        
```

```python
# Save the stock_market_data in csv files
save_stock_market_data(datetime.datetime(2010, 1, 1))
```

The data that isn't found will be discussed in the report. In short, some of these companies are listed under more than one ticker due to Class A and Class C shares. Moreover, some companies have been restructured and broken up. To keep consistency over the training dates, I excluded companies that have been restrcutured. These form a negligible proportion of the S&P 500.
